```python
from transformers import pipeline

# Load the Question-Answering pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Example document
document = """
Artificial Intelligence (AI) is transforming the way businesses operate, enhancing
efficiency and decision-making processes. This technology is widely adopted across
various industries, including healthcare, finance, and transportation, due to its
ability to analyze large datasets quickly and accurately. However, ethical concerns
surrounding AI's use, such as data privacy and algorithmic bias, remain pressing issues.
"""

# Questions to ask
questions = [
    "What industries use AI?",
    "What benefits does AI provide?",
    "What are the ethical concerns surrounding AI?"
]

# Analyze the document with Q&A
print("Document Analysis - Q&A Section")
for question in questions:
    result = qa_pipeline(question=question, context=document)
    print(f"\nQuestion: {question}")
    print(f"Answer: {result['answer']}")
```

##Sentence Similarity
Task: Computes similarity between two sentences.
##Common Models:
sentence-transformers/all-MiniLM-L6-v2

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from transformers import pipeline

# Step 1: Initialize the pipeline
similarity = pipeline("feature-extraction", model="sentence-transformers/all-MiniLM-L6-v2")

# Step 2: Extract embeddings for sentences
embeddings = similarity(["This is a sentence.", "This is another sentence."])

# Debug: Check original shape
print(f"Original embeddings shape: {np.array(embeddings).shape}")  # (2, 1, 7, 384)

# Step 3: Remove batch dimension and apply mean pooling
# Squeeze removes the batch dimension (1), and mean pooling reduces over sequence length (7)
sentence_embeddings = [np.mean(np.squeeze(embed), axis=0) for embed in embeddings]

# Debug: Check shape after pooling
print(f"Sentence embeddings shape: {np.array(sentence_embeddings).shape}")  # (2, 384)

# Step 4: Compute cosine similarity
similarity_score = cosine_similarity([sentence_embeddings[0]], [sentence_embeddings[1]])
print(f"Cosine Similarity: {similarity_score[0][0]}")
```
